```python
import gradio as gr
import pandas as pd
import joblib
```

```python
modelo = joblib.load('modelo_colesterol.pkl')
```

```python
def predict(grupo_sanguineo, fumante, nivel_atividade_fisica, idade, peso, altura):
    input_data = pd.DataFrame({
        'grupo_sanguineo': [grupo_sanguineo],
        'fumante': [fumante],
        'nivel_atividade_fisica': [nivel_atividade_fisica],
        'idade': [idade],
        'peso': [peso],
        'altura': [altura]
    })

    prediction = modelo.predict(input_data)
    
    return prediction[0]
```

```python
iface = gr.Interface(
    fn=predict,
    inputs=[
        gr.Dropdown(choices=['A', 'B', 'AB', 'O'], label='Grupo Sanguíneo', value='O'),
        gr.Dropdown(choices=['Sim', 'Não'], label='Fumante', value='Não'),
        gr.Dropdown(choices=['Baixo', 'Moderado', 'Alto'], label='Nível de Atividade Física', value='Moderado'),
        gr.Slider(minimum=20, maximum=120, step=1, label='Idade'),
        gr.Slider(minimum=30, maximum=200, step=1, label='Peso (kg)'),
        gr.Slider(minimum=100, maximum=250, step=1, label='Altura (cm)')
    ],
    outputs=gr.Number(label='Nível de Colesterol Predito'),
    title='Predição de Nível de Colesterol',
    description='Insira os dados do paciente para prever o nível de colesterol.'
)

iface.launch(share=True)
```
